## Closest to levelling up

### Here is a data dragon on champ properties and stuff

```python
ddragon = requests.get('http://ddragon.leagueoflegends.com/cdn/11.11.1/data/en_US/champion.json')
```

```python
champData = ddragon.json()
```

```python
champData
```

```python
champData['data']
```

```python
NameToId = {}
for champ in champData['data']:
    NameToId[champData['data'][champ]['key']] = champData['data'][champ]['id']
print(NameToId)
```

```python
PUNLDic = []
for champion in mscores:
    if champion['championPointsUntilNextLevel'] != 0:
        name = NameToId[str(champion['championId'])]
        PUNLDic.append({'Name': name, 'punl': champion['championPointsUntilNextLevel']})
PUNLDic.sort(key=lambda x: x.get('punl'))
print(PUNLDic)
```

```python
for champ in PUNLDic:
    print(champ['Name'] + ": " + str(champ['punl']))
```

# Game Stats

```python
#however many games ago here
hmga = '0'
lastGame = requests.get('https://americas.api.riotgames.com/lol/match/v5/matches/by-puuid/' + puuId + '/ids'+ '?api_key=' + apiKey + '&start=' +hmga + '&count=1')
lg = lastGame.json()[0]
matchData = requests.get('https://americas.api.riotgames.com/lol/match/v5/matches/' + lg + '?api_key=' + apiKey)
md = matchData.json()
md
```

```python
lg
```

```python
gmode = {'CLASSIC': 'Summoner\'s Rift (Normals)','ARAM': 'Howling Abyss (ARAM)','NEXUSBLITZ': 'Nexus Blitz'}
gtime = md['info']['gameDuration']
print('Gamemode: ' + gmode[md['info']['gameMode']])
print('Duration: ' + str(gtime//60000) + 'm' + str((gtime-60000*(gtime//60000))//1000) + 's')
```

```python
index = md['metadata']['participants'].index(puuId)
sumStats = md['info']['participants'][index]
sumStats
```

```python
mkconversion = {1:'Single kill',2:'Double kill',3: 'Triple kill',4:'Quadra kill',5:'Pentakill'}
winconversion = {True:'Won',False:'Lost'}

kda = str(sumStats['kills']) + '/' + str(sumStats['deaths']) + '/' + str(sumStats['assists'])
champName = sumStats['championName']
position = sumStats['individualPosition'].lower().title()
largestmk = mkconversion[sumStats['largestMultiKill']]
bint = str(sumStats['totalTimeSpentDead']//60) + 'm' + str(sumStats['totalTimeSpentDead']%60) + 's' + '(' + '{:.2f}'.format(sumStats['totalTimeSpentDead']/sumStats['timePlayed']*100) + '%)'
win = winconversion[sumStats['win']]

damageDealt = str(sumStats['totalDamageDealtToChampions'])
damageTaken = str(sumStats['totalDamageTaken'])
damageMitigated = str(sumStats['damageSelfMitigated'])
damageToObjectives= str(sumStats['damageDealtToObjectives'])
healing = str(sumStats['totalHeal'])

parti = md['info']['participants']

dD =[]
for player in parti:
    dD.append(player['totalDamageDealtToChampions'])
dD = sorted(dD,reverse=True)
dDr = 'unknown'
for i in range(1,11):
    if dD[i-1] == sumStats['totalDamageDealtToChampions']:
        dDr = str(i)

dT =[]
for player in parti:
    dT.append(player['totalDamageTaken'])
dT = sorted(dT,reverse=True)
dTr = 'unknown'
for i in range(1,11):
    if dT[i-1] == sumStats['totalDamageTaken']:
        dTr = str(i)

dM =[]
for player in parti:
    dM.append(player['damageSelfMitigated'])
dM = sorted(dM,reverse=True)
dMr = 'unknown'
for i in range(1,11):
    if dM[i-1] == sumStats['damageSelfMitigated']:
        dMr = str(i)

dO =[]
for player in parti:
    dO.append(player['damageDealtToObjectives'])
dO = sorted(dO,reverse=True)
dOr = 'unknown'
for i in range(1,11):
    if dO[i-1] == sumStats['damageDealtToObjectives']:
        dOr = str(i)

tH =[]
for player in parti:
    tH.append(player['totalHeal'])
tH = sorted(tH,reverse=True)
tHr = 'unknown'
for i in range(1,11):
    if tH[i-1] == sumStats['totalHeal']:
        tHr = str(i)
```

```python
print('Gamemode: ' + gmode[md['info']['gameMode']])
print('Duration: ' + str(gtime//60000) + 'm' + str((gtime-60000*(gtime//60000))//1000) + 's')

print('Outcome: ' + win)
print('Champion: ' + champName)
print('Role: ' + position)
print('KDA: ' + kda)
print('Largest Multikill: ' + largestmk)
print('Time Binting: ' + bint)

print('Damage Dealt: ' + damageDealt + ' (' + dDr + ')')
print('Damage Taken: ' + damageTaken + ' (' + dTr + ')')
print('Damage Mitigated: ' + damageMitigated + ' (' + dMr + ')')
print('Damage to Objectives: ' + damageToObjectives + ' (' + dOr + ')')
print('Healing: ' + healing + ' (' + tHr + ')')
```
