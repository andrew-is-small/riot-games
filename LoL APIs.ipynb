# Closest to levelling up

```python
ddragon = requests.get('http://ddragon.leagueoflegends.com/cdn/11.11.1/data/en_US/champion.json')
```

```python
champData = ddragon.json()
```

```python
champData['data']
```

```python
NameToId = {}
for champ in champData['data']:
    NameToId[champData['data'][champ]['key']] = champData['data'][champ]['id']
print(NameToId)
```

```python
PUNLDic = []
for champion in mscores:
    if champion['championPointsUntilNextLevel'] != 0:
        name = NameToId[str(champion['championId'])]
        PUNLDic.append({'Name': name, 'punl': champion['championPointsUntilNextLevel']})
PUNLDic.sort(key=lambda x: x.get('punl'))
print(PUNLDic)
```

```python
for champ in PUNLDic:
    print(champ['Name'] + ": " + str(champ['punl']))
```

# Game Stats

```python
#however many games ago here
hmga = '2'
lastGame = requests.get('https://americas.api.riotgames.com/lol/match/v5/matches/by-puuid/' + puuId + '/ids'+ '?api_key=' + apiKey + '&start=' +hmga + '&count=1')
lg = lastGame.json()[0]
matchData = requests.get('https://americas.api.riotgames.com/lol/match/v5/matches/' + lg + '?api_key=' + apiKey)
md = matchData.json()
md
```

```python
gmode = {'CLASSIC': 'Summoner\'s Rift (Normals)','ARAM': 'Howling Abyss (ARAM)'}
gtime = md['info']['gameDuration']
print('Gamemode: ' + gmode[md['info']['gameMode']])
print('Duration: ' + str(gtime//60000) + 'm' + str((gtime-60000*(gtime//60000))//1000) + 's')
```

```python
index = md['metadata']['participants'].index(puuId)
sumStats = md['info']['participants'][index]
sumStats
```

```python
mkconversion = {1:'Single kill',2:'Double kill',3: 'Triple kill',4:'Quadra kill',5:'Pentakill'}
winconversion = {True:'Won',False:'Lost'}

kda = str(sumStats['kills']) + '/' + str(sumStats['deaths']) + '/' + str(sumStats['assists'])
champName = sumStats['championName']
position = sumStats['individualPosition'].lower().title()
largestmk = mkconversion[sumStats['largestMultiKill']]
bint = str(sumStats['totalTimeSpentDead']) + 's (' + '{:.2f}'.format(sumStats['totalTimeSpentDead']/sumStats['timePlayed']*100) + '%)'
win = winconversion[sumStats['win']]
```

```python
print('Gamemode: ' + gmode[md['info']['gameMode']])
print('Duration: ' + str(gtime//60000) + 'm' + str((gtime-60000*(gtime//60000))//1000) + 's')

print('Outcome: ' + win)
print('Champion: ' + champName)
print('Role: ' + position)
print('KDA: ' + kda)
print('Largest Multikill: ' + largestmk)
print('Time Binting: ' + bint)
```
